```python
import wmfdata as wmf
```

```python
contributor_event = wmf.presto.run("""
SELECT
    CAST(FROM_ISO8601_TIMESTAMP(meta.dt) AS DATE) AS dt,
    COUNT(*) AS events
FROM event.mediawiki_wikistories_contribution_event
WHERE
    year >= 0
GROUP BY
    CAST(FROM_ISO8601_TIMESTAMP(meta.dt) AS DATE)
ORDER BY
    dt ASC
""").set_index("dt")
```

```python
contributor_event
```

```python
sanitized_contributor_event = wmf.presto.run("""
SELECT
    CAST(FROM_ISO8601_TIMESTAMP(meta.dt) AS DATE) AS dt,
    COUNT(*) AS events
FROM event_sanitized.mediawiki_wikistories_contribution_event
WHERE
    year >= 0
GROUP BY
    CAST(FROM_ISO8601_TIMESTAMP(meta.dt) AS DATE)
ORDER BY
    dt ASC
""").set_index("dt")
```

This looks a bit weird: we have data from 2022-08-21 to 2022-08-28, and then again from 2022-10-04 to 2022-10-12. But that actually looks right, because there are two sanitization jobs, one running in real time and the other after 45 days. We started sanitizing on 2022-10-04, and the follow-up job is gradually filling in the old data.

```python
sanitized_contributor_event
```
